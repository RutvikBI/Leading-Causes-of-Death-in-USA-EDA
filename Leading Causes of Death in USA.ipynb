### Are Americans facing increasing, decreasing, or steady likelihood of death?

```python
import pandas as pd
import numpy as np
#read csv with death causes
deathcause_data=pd.read_csv('C:/Users/Rushabh-Pc/Desktop/Fall 2019/python bana/NCHS_-_Leading_Causes_of_Death__United_States.csv')
deathcause_data.head(2)
```

```python
#read excel with census data; notice that the data available is from 2010 to 2018
population_data=pd.read_excel('C:/Users/Rushabh-Pc/Desktop/Fall 2019/python bana/nst-est2018-01.xlsx',header=3,index_col=0)
population_data.head(2)
```

```python
cols_year_deaths=[0,4]
yearly_deaths = deathcause_data[deathcause_data.columns[cols_year_deaths]]
#notice that data available is from 1999 to 2016
grouped_years = yearly_deaths.groupby('Year').sum().T
grouped_years
```

```python
#common years from both tables
available_year_data = list(grouped_years.columns.intersection(population_data.columns))
available_year_data
```

```python
#In the generic form, mortality rates are calculated as: deaths/population*10^3  ,for per 1000 population
death_rate=(grouped_years[available_year_data]/population_data[available_year_data].loc['United States'])*1000
print("Deaths per 1000 Americans has not drastically changed over the year and is as follows:")
death_rate
```

### What are the four leading causes of death for Americans?

```python
cols_cause_deaths=[2,4]
cause_deaths = deathcause_data[deathcause_data.columns[cols_cause_deaths]]
grouped_cause = cause_deaths.groupby('Cause Name').sum()
four_leading_causes=grouped_cause.sort_values(by=['Deaths'],ascending=False).iloc[1:5]
print('Four leading causes of death for Americans:')
four_leading_causes
```

### Do individual states show the same four leading causes of death?

```python
#notice that data available is from 1999 to 2016
grouped_state = deathcause_data[['State','Cause Name','Deaths']].groupby(['State','Cause Name']).aggregate(sum).sort_values(by=['Deaths'],ascending=False).reset_index('State')
grouped_state=grouped_state
state_four_leading_causes=grouped_state.groupby('State').apply(lambda x:x.iloc[1:5]).reset_index('Cause Name')
#disp_state_four_leading_causes=state_four_leading_causes.reset_index(drop='State')
#disp_state_four_leading_causes['Cause Name'].value_counts().head(4)
#The Frequency counts of contents in the Cause Name column display the same 4 Causes
lead_4_causes_state=state_four_leading_causes['Cause Name'].value_counts()
lead_4_causes_state

answer=lead_4_causes_state/52*100
print(lead_4_causes_state,"\n",answer)
```

### Are there year-by-year changes in the four leading causes of death nationwide?

```python
grouped_year = deathcause_data[['Year','Cause Name','Deaths']].groupby(['Year','Cause Name']).aggregate(sum).sort_values(by=['Deaths'],ascending=False).reset_index('Year')
grouped_year=grouped_year
year_four_leading_causes=grouped_year.groupby('Year').apply(lambda x:x.iloc[1:5]).reset_index('Cause Name')
year_four_leading_causes
lead_causes_year=year_four_leading_causes['Cause Name'].value_counts()
lead_causes_year
answer2=lead_causes_year/18*100
print(year_four_leading_causes)
print(lead_causes_year,'\n',answer2)
```
